- Erzeuge fünf Sinussignale im Interval $[0, 6 \pi), bestehend aus 200 Datenpunkten
- Addiere jeweils Gauss-verteiltes Rauschen mit einem Vorfaktor von 0.1
- Setze den Mittelwert der Signale auf 0, 2, 4, 6, 8

```python
# Erzeuge ein Sinussignal
sin_signal = np.sin(np.linspace(0, 6 * np.pi, num=200))

# Erzeuge 5 x Rauschen mit 200 Datenpunkten
noise = np.random.randn(200, 5)

# Addiere das Rauschen via Broadcasting -> 5 verrauschte Sinussignale
noisy_signals = sin_signal[:, np.newaxis] + 0.1 * noise

# Addiere den jeweiligen Mittelwert via Broadcasting
mean_noisy_signal = noisy_signals + np.arange(0, 10, 2)

plot(mean_noisy_signal);
```

- Renormiere die Signale auf [-1, 1]

```python
# 1. Mittelwert auf 0 setzen:
demeaned_signal = mean_noisy_signal - mean_noisy_signal.mean(axis=0)

# 2. Renormierung:
new_signal = (demeaned_signal / (demeaned_signal.max(axis=0) - demeaned_signal.min(axis=0))) * 2
```

```python
plot(new_signal);
```
