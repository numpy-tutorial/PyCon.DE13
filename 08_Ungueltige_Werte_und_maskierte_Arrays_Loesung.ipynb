```python
import numpy as np
```

```python
import numpy.ma as ma
```

```python
lena = np.load('files/invalid_array.npy')
```

```python
lena = ma.masked_invalid(lena)
```

```python
lena.mean()
```

```python
lena = ma.masked_less(lena, -50)
```

```python
lena.mean()
```

```python
lena.sum(axis=1)
```

```python
mask = lena.mask
```

```python
lena.data[mask.sum(axis=1)]
```
